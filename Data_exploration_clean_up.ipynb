```python
# extracting only europe data

data_europe = data[
    (data['latitude'] >= 20) &
    (data['latitude'] <= 80) &
    (data['longitude'] >= -50) &
    (data['longitude'] <= 50)]
```

```python
#separating america and australia to employ the model

data_america = data[
    (data['latitude'] >= -40) &
    (data['latitude'] <= 80) &
    (data['longitude'] >= -150) &
    (data['longitude'] <= -50)]

data_australia = data[
    (data['latitude'] >= -50) &
    (data['latitude'] <= -10) &
    (data['longitude'] >= 100) &
    (data['longitude'] <= 150)]
```

```python
#saving data cleaned

# Creating folder
folder = Path("processed_data")
folder.mkdir(exist_ok=True)
```

```python
data.to_csv(folder / "data.csv", index=False)
data_europe.to_csv(folder / "data_europe.csv", index=False)
data_america.to_csv(folder / "data_america.csv", index=False)
data_australia.to_csv(folder / "data_australia.csv", index=False)
```
