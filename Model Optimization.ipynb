```python
#importing libraries

import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from lightgbm import LGBMRegressor
from sklearn.model_selection import KFold, cross_val_score
```

```python
#import data

data = pd.read_csv('processed_data/data.csv')

#separate 20% of data to speed up test

data_20 = data.sample(frac=0.2, random_state=1)
```

```python
data.isna().sum()
```

```python
#split training and data
features = data_20.select_dtypes(include='number')
X = features.drop(columns='price')
y = data_20['price']

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=1)
```

```python
features.head()
```

```python
#dictionary to record result

results = {}
```

```python
#teasting Linear regressor
#define_model
lr = LinearRegression()
#training_model
lr.fit(X_train,y_train)
#predicting
y_pred = lr.predict(X_test)

#evaluating model

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)


results['Linear Regression'] = {
    'MAE': mae,
    'RMSE': rmse,
    'R2': r2
}
```

```python
print(results)
```

```python
#teasting Decision Tree
#define_model
DT = DecisionTreeRegressor(random_state=1)
#training_model
DT.fit(X_train,y_train)
#predicting
y_pred = DT.predict(X_test)

#evaluating model

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)


results['Decision Tree'] = {
    'MAE': mae,
    'RMSE': rmse,
    'R2': r2
}
```

```python
print(results)
```

```python
#teasting RandomForestRegressor
#define_model
RF = RandomForestRegressor(
    n_estimators=100,
    random_state=1)
#training_model
RF.fit(X_train,y_train)
#predicting
y_pred = RF.predict(X_test)

#evaluating model

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)


results['Random Forest'] = {
    'MAE': mae,
    'RMSE': rmse,
    'R2': r2
}
```

```python
print(results)
```

```python
#teasting Gradient Boosting Regressor
#define_model
GB = GradientBoostingRegressor(random_state=1)
#training_model
GB.fit(X_train,y_train)
#predicting
y_pred = GB.predict(X_test)

#evaluating model

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)


results['Gradient Boosting Regressor'] = {
    'MAE': mae,
    'RMSE': rmse,
    'R2': r2
}
```

```python
print(results)
```

```python
#teasting Ligth GBM Regressor
#define_model
LGBM = LGBMRegressor(random_state=1)
#training_model
LGBM.fit(X_train,y_train)
#predicting
y_pred = LGBM.predict(X_test)

#evaluating model

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)


results['Ligth GBM Regressor'] = {
    'MAE': mae,
    'RMSE': rmse,
    'R2': r2
}
```

As expected, geographical features, especially longitude and latitude, were the most important variables in the model, highlighting the strong influence of location on Airbnb prices. However, the feature importance analysis also revealed that id and host_id contributed significantly to the model prediction. Since these variables are identifiers rather than meaningful characteristics of the properties, their importance may indicate the presence of data leakage or hidden patterns associated with the listing registration process. Therefore, these features will be removed in the following steps, and the model will be retrained to evaluate its real predictive capability.

Latitude and longitude will be maintained, as they represent meaningful geographical information that can capture differences in location and market value.

```python
model = LGBMRegressor(
    n_estimators=500,
    learning_rate=0.1,
    num_leaves=64,
    max_depth=8,
    min_child_samples=50,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=1
)

cv = KFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

scores = cross_val_score(
    model,
    X_train_full,
    y_train_full,
    cv=cv,
    scoring="neg_root_mean_squared_error"
)

rmse = -scores

```

```python
print("RMSE each fold:", rmse)
print("RMSE mean:", rmse.mean())
print("Standard Deviation:", rmse.std())
```
